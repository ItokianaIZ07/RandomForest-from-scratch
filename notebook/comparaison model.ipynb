```python
df = pd.read_csv('../data/data.csv', sep=";")
df.head()
```

```python
arbreMaison = DecisionTree(10)
foretMaison = RandomForest(10)
arbreSKL = DecisionTreeClassifier(criterion='gini', max_depth=10)
foretSKL = RandomForestClassifier(n_estimators=10)
```

```python
X = df.drop(columns=["est_malade"])
y = df["est_malade"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
models = [arbreMaison, arbreSKL, foretMaison, foretSKL]
times = []
for model in models:
    start = time.time()
    model.fit(X_train, y_train)
    end = time.time()
    times.append(end-start)
```

```python
print(times)
```

```python
for model in models:
    y_pred = model.predict(X_test)
    f1 = f1_score(y_test, y_pred)
    print(f1)
```
